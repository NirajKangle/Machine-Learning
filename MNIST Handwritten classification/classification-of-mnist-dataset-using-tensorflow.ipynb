```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import tensorflow as tf
from tensorflow import keras
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train), 28*28)
X_test_flattened = X_test.reshape(len(X_test), 28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape = (784,), activation = 'sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_flattened, y_train, epochs = 5)
```

```python
model.evaluate(X_test_flattened, y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_pred = model.predict(X_test_flattened)
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
y_pred_labels[:5]
```

```python
y_test[:5]
```

```python
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_pred_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize = (12,6))
sn.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('true')
```

```python
model2 = keras.Sequential([
    keras.layers.Dense(100, input_shape = (784,), activation = 'relu'),
    keras.layers.Dense(10, activation = 'sigmoid') #No need to add input features because it will figure it out baased on original input
])

model2.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model2.fit(X_train_flattened, y_train, epochs = 5)
```

```python
model2.evaluate(X_test_flattened, y_test)
```

```python
y_pred = model2.predict(X_test_flattened)
y_pred_labels = [np.argmax(i) for i in y_pred]
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_pred_labels)

plt.figure(figsize = (12,6))
sn.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('true')
```

```python
model3 = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation = 'relu'), #No need to add input features because it will figure it out baased on original input
    keras.layers.Dense(10, activation = 'sigmoid') #No need to add input features because it will figure it out baased on original input
])

model3.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model3.fit(X_train, y_train, epochs = 5)
```

```python
y_pred = model3.predict(X_test_flattened)
y_pred_labels = [np.argmax(i) for i in y_pred]
cm = tf.math.confusion_matrix(labels = y_test, predictions = y_pred_labels)

plt.figure(figsize = (12,6))
sn.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('predicted')
plt.ylabel('true')
```
